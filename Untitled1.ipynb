```python
pip install pyaudio
pip install SpeechRecognition
import speech_recognition as sr
import os
import pyaudio
import wave
from google.cloud import speech_v1p1beta1 as speech

# Ses tanıma motorunu başlat
r = sr.Recognizer()

# Mikrofonu başlat
with sr.Microphone() as source:
    #Sistem mikrofonu burada açıldı ve hey kapsül cümlesinin gelmesini bekliyo

    r.adjust_for_ambient_noise(source)  # Arka plan gürültüsünü dikkate al
    while True:
        audio = r.listen(source)  # Ses kaydını yap

        try:
            # Ses kaydını metne dönüştür
            text = r.recognize_google(audio, language="tr-TR")  # Türkçe dil tanıma

            # Algılanan metinde "Hey kapsül" kelimesini ara
            if "hey kapsül" in text.lower():
                

                # Kayıt ayarları
                FORMAT = pyaudio.paInt16
                CHANNELS = 1
                RATE = 44100
                CHUNK = 1024
                RECORD_SECONDS = 7

                # Masaüstü dizinini alın
                desktop_path = os.path.join(os.path.expanduser("~"), "Desktop")

                # Kaydedilecek dosyanın tam yolunu oluşturun
                WAVE_OUTPUT_FILENAME = os.path.join(desktop_path, "output.wav")

                # PyAudio nesnesini başlat
                audio = pyaudio.PyAudio()

                # Ses akışını başlat
                stream = audio.open(format=FORMAT, channels=CHANNELS, rate=RATE, input=True, frames_per_buffer=CHUNK)

                print("Merhaba! Size nasıl yardımcı olabilirim?")

                frames = []

                # Ses verilerini topla
                for i in range(0, int(RATE / CHUNK * RECORD_SECONDS)):
                    data = stream.read(CHUNK)
                    frames.append(data)


                # Ses akışını durdur ve kapat
                stream.stop_stream()
                stream.close()
                audio.terminate()

                # Ses verilerini WAV dosyasına kaydet
                wf = wave.open(WAVE_OUTPUT_FILENAME, 'wb')
                wf.setnchannels(CHANNELS)
                wf.setsampwidth(audio.get_sample_size(FORMAT))
                wf.setframerate(RATE)
                wf.writeframes(b''.join(frames))
                wf.close()

                

                # Google Cloud kimlik bilgilerini ayarlayın
                os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = r"C:\Users\kayai\OneDrive\Masaüstü\workspace(K)\Ses Analizi\credentials.json"

                # Ses dosyasının yolu
                AUDIO_FILE = WAVE_OUTPUT_FILENAME

                # Konuşma hizmeti istemcisini oluşturun
                client = speech.SpeechClient()

                # Ses dosyasını yükleme
                with open(AUDIO_FILE, "rb") as audio_file:
                    content = audio_file.read()

                audio = speech.RecognitionAudio(content=content)

                # Sesin dili ve formatı
                config = speech.RecognitionConfig(
                    encoding=speech.RecognitionConfig.AudioEncoding.LINEAR16,
                    sample_rate_hertz=44100,
                    language_code="tr-TR"
                )

                # Konuşmayı metne dönüştürme
                response = client.recognize(config=config, audio=audio)

                # transcript içindeki metin chatbot içine gönderilecek ve chatgpt transcript içindeki cümleye cevap verecek.
                
                transcript = ""
                for result in response.results:
                    transcript += result.alternatives[0].transcript

                # Metni yazdır
                print("Metin: ", transcript)

        except sr.UnknownValueError:
            print("Ses anlaşılamadı. Lütfen tekrar deneyin.")
        except sr.RequestError as e:
            print("Hata oluştu; {0}".format(e))
```
